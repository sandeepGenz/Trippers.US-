```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv('tourism.csv')
df
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.concat([df,pd.get_dummies(df['AttractionType'])],axis=1)
```

```python
df
```

```python
df.drop(['AttractionType'],axis=1,inplace=True)
df

# "AttractionsNearby":20	"RestaurantsNearby":5	"Adventure":1	"Amusement":0	"Historical":0	"Shopping":0	"Worship":0
```

```python

X = df.drop(['KidFriendlyscore','LowMobilityscore','SeniorCitizenscore'],axis=1).values
X
```

```python
y = df[['KidFriendlyscore','LowMobilityscore','SeniorCitizenscore']].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
X_train, X_test, y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
print(model.predict([[4.23, 1.16, 0, 1,0, 0, 0]]))
print((np.round(model.predict([[4.23, 1.16, 0, 1,0, 0, 0]]))).astype(int))
```

```python
from sklearn.metrics import classification_report
y_pred = (np.round(model.predict(X_test))).astype(int)
for i in range(len(y_test)):
  print(y_test[i], y_pred[i])
```

```python
from numpy import absolute, mean, std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X_test, y_test, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
n_scores = absolute(n_scores)
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
